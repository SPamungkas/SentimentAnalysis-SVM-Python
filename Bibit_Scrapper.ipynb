```python
from google_play_scraper import app

import pandas as pd

import numpy as np
```

```python
# Scrape all available reviews
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.bibit.bibitid',
    lang='id', #default 'en'
    country='id', #default to us
    sort=Sort.MOST_RELEVANT, #default sort most relevant
    count=2000, #defaults 100
    filter_score_with=None #defaults to none use 1,2,3,4,5 to select score
) 
```

```python
bibit_raw =  pd.DataFrame(np.array(result),columns=['review'])

bibit_raw = bibit_raw.join(pd.DataFrame(bibit_raw.pop('review').tolist()))

bibit_raw.head()
```

```python
len(bibit_raw.index)
```

```python
# labelling

label = []
for index, row in bibit_raw.iterrows():
  if row["score"] == 5 or row["score"] == 4:
    label.append(1)
  else:
    label.append(0)

bibit_raw["label"] = label
bibit_raw
```

```python
bibit_raw.to_csv("bibit_label.csv", index = False)
```
